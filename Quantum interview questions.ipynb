# Q1.A sorted array is rotated at some unknown point, find the minimum element in
# it.

```python
## Q1.A sorted array is rotated at some unknown point, find the minimum element in it.

import numpy as np
import random
c=int(input('enter starting value of array range :- '))
j=int(input('enter ending value of array range :- '))
k=int(input('how many random nos you want to print:- '))
array=np.random.randint(c,j,k)
print(array)
rot=int(input('how many time you want to rotate an array: '))
ls=list(array)
print(ls)
for i in range(rot):
    first=ls[0]
    print(first)
    ls.remove(ls[0])
    ls.append(first)
    
print(f'the {rot} times rotated array is: ',ls)
print(f'The minimum value in the array range is:',min(ls)) 
```

# convert ‘str1′ into ‘str2′.

```python
str1=str(input('enter correct word: '))
print(str1)
str2=str(input('enter same word with wrong letters: '))
print(str2)
c=0
ar=[]
for i in range(len(str1)):
    if str2[i] != str1[i]:
        c+=1
        ar.append(str1[i])
    elif str2[i]==str1[i]:
        ar.append(str1[i])

res = ''.join(ar)
print('the replaced characters word: ',str2,': ',ar)
print('the count of replaced characters:- ',c)
```
